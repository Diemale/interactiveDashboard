## Create a 'countries' column from 'crash_location'

```python
df2['countries'] = ''    # we create a new column that will storage countries names.
```

```python
# The function below receives a string as a parameter. If the string contains a comma the function will split it on the last 
# comma. Then it'll strip it, and save the second part of the text, where, we hope, there will be a country's name. Otherwise, 
# it'll return the original string.
```

```python
def rsplitString(STR): 
    if ',' in STR:
        lRS = STR.rsplit(',', 1)
        stripped = lRS[1].strip()    
    else:
        stripped = STR
    return stripped
```

```python
# Now we'll apply our function to 'crash_location'
```

```python
df2["countries"] = df2["crash_location"].apply(lambda x: rsplitString(x))
```

```python
df3 = df2.copy()
```

```python
# Well, it didn't result as we expected, but with a few adjustments, we'll get there.
# Some of the unexpected results are: In the case of the United States, our column received states' names instead of 
# "United States". Something similar happened with the United Kingdom besides the name was listed as UK (in some cases), not
# the complete name. Some registers have unwanted words preceding the country's name. 
# Let's work around these issues.
```

```python
# A function to remove undesired leading words.
def remove_lwords(STR):
    palabrasMolestas = [ 'Over the Gulf of ', 'the Gulf of ','Off the coast of ', 'Jacquinot Bay ', 
                       'Forest-in-Teesdale. ', 'Gulf of ', 'South of ', 'Near ',
                       'NE of ', 'Territory of ','Irkutsk ','Qld. ', 'Off ','off ','Over ','French '
                       ]
    
    cleaned= ''
    for word in palabrasMolestas:
        if STR.startswith(word):
            cleaned = STR.replace(word, "")
            return cleaned  
    return STR
```

```python
df3["countries"] = df3["countries"].apply(remove_lwords)
```

## Make a collisions column

```python
# We noticed that some cells got more than a single data in them, which could mean that there was more than one airplane 
# involved in the accident. So, let's find out if we are right. To narrow the chances of making a mistake, we'll apply the 
# function separators to two columns instead of one to get a clue about which cells have composite data. Finally, we'll  
# filter the result and create a DataFrame with this data.
```

```python
def separators(cad):
    isIn = False
    
    if (cad.find('/') != -1) | (cad.find('&') != -1): 
        isIn =  True
    return isIn    
```
